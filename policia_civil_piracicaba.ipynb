```python
import pandas as pd
import matplotlib.pyplot as plt

arquivo = 'produtividade_policial_piracicaba_2010_2021.csv'
df = pd.read_csv(arquivo).drop('Unnamed: 0', axis='columns')
df = df.fillna(0)
df
```

```python
criterio = (df['Natureza']=='Nº DE FLAGRANTES LAVRADOS')|(df['Natureza']=='TOT. DE INQUÉRITOS POLICIAIS INSTAURADOS')
```

```python
df = df[criterio]
df = (df[['Ano', 'Delegacia', 'Natureza', 'Total']])
df
```

```python
grouped = df.groupby(['Ano', 'Natureza', 'Delegacia'])['Total'].sum()
print (grouped.to_string()) # Para exibir todas as linhas da Series
```

```python
grouped1 = df.groupby(['Ano', 'Natureza'])['Total'].sum()
display(grouped1)
```

```python
criterio1 = (df['Natureza']=='Nº DE FLAGRANTES LAVRADOS')
grouped2 = df[criterio1].groupby('Ano')['Total'].sum()
print ('Total de flagrantes lavrados')
print(grouped2)
```

```python
print('Estatística de flagrantes lavrados')
grouped2.describe()
```

```python
criterio2 = (df['Natureza']=='TOT. DE INQUÉRITOS POLICIAIS INSTAURADOS')
grouped3 = df[criterio2].groupby('Ano')['Total'].sum()
print ('Total de inquéritos instaurados')
print(grouped3)
```

```python
print('Estatística de inquéritos instaurados')
grouped3.describe()
```

```python
delegacias = df['Delegacia'].unique()
delegacias
```

```python
for n in delegacias:
  criterio3 = (df['Natureza'] == 'Nº DE FLAGRANTES LAVRADOS') & (df['Delegacia'] == n)
  fig = plt.figure()
  axes1 = fig.add_subplot(1, 1, 1)
  axes1.set_title(f'Flagrantes lavrados p/ {n}')
  axes1.bar(df[criterio3]['Ano'], df[criterio3]['Total'])
```
